# Connecting to MySQL DB

Make sure to have SQL Alchemy pip installed and to use the config.py file from Slack (or Google Docs)

The config.py file should have:
- db_user = 'XXXXXXX'
- db_pass = 'XXXXXXX'
- db_name = 'XXXXXXX'
- db_host = 'XXXXXXX'
- db_main = 'XXXXXXX'

db_name is the original landing database, db_main is the db we will have the cleaner data with cluster ref

```python
import pandas as pd
from datetime import datetime
import numpy as np

import config

from sqlalchemy import create_engine
```

```python
# create sqlalchemy engine
engine = create_engine("mysql+mysqlconnector://{user}:{password}@{host}/{dbname}"
                       .format(user=config.db_user,
                               password=config.db_pass,
                               dbname=config.db_main,
                               host=config.db_host))
```

```python
query1 = """
SELECT *
FROM Weather;
"""
```

```python
query2 = """
SELECT *
FROM Emissions_Data
LIMIT 10000;
"""
```

```python
df1 = pd.read_sql(query1, engine)
df2 = pd.read_sql(query2, engine)
```

```python
engine.dispose()
```
